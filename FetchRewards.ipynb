```python
import pandas as pd
import json
from pandas.io.json import json_normalize
from datetime import datetime
```

```python
# Load json data and normalize JSON data into a flat table
receipts_json = [json.loads(line) for line in open('D:/users/FR/receipts.json', 'r')]
receipts = json_normalize(receipts_json)

users_json = [json.loads(line) for line in open('D:/users/FR/users.json', 'r')]
users = json_normalize(users_json)

brands_json = [json.loads(line) for line in open('D:/users/FR/brands.json', 'r')]
brands = json_normalize(brands_json)
```

```python
# Review users dataset
users
```

```python
# Check for missing value
users.isnull().sum()
```

```python
# Update column names
users.rename(columns={'_id.$oid': 'id', 'createdDate.$date': 'createdDate', 'lastLogin.$date': 'lastLogin'}, inplace=True)

# Convert unix timestamp to human readble timestamp
users['createdDate'] = pd.to_datetime(users['createdDate'], errors='ignore', unit='ms', origin='unix')
users['lastLogin'] = pd.to_datetime(users['lastLogin'], errors='ignore', unit='ms', origin='unix')

# Remove duplicate rows
users.drop_duplicates(inplace=True)
```

```python
# Review brands dataset
brands
```

```python
# Check for missing value
brands.isnull().sum()
```

```python
# Update column names
brands.rename(columns={'_id.$oid': 'id', 'cpg.$id.$oid': 'cpg_id', 'cpg.$ref': 'cpg_ref'}, inplace=True)
brands
```

```python
# Review receipts dataset
receipts
```

```python
# Check for missing value
receipts.isnull().sum()
```

```python
# Update column names
receipts.rename(columns={'_id.$oid': 'id', 'createDate.$date': 'createDate', 'dateScanned.$date': 'dateScanned', 'finishedDate.$date' : 'finishedDate', 'modifyDate.$date' : 'modifyDate', 'pointsAwardedDate.$date' : 'pointsAwardedDate', 'purchaseDate.$date': 'purchaseDate'}, inplace=True)

# Convert unix timestamp to human readble timestamp
receipts['createDate'] = pd.to_datetime(receipts['createDate'], errors='ignore', unit='ms', origin='unix')
receipts['dateScanned'] = pd.to_datetime(receipts['dateScanned'], errors='ignore', unit='ms', origin='unix')
receipts['finishedDate'] = pd.to_datetime(receipts['finishedDate'], errors='ignore', unit='ms', origin='unix')
receipts['modifyDate'] = pd.to_datetime(receipts['modifyDate'], errors='ignore', unit='ms', origin='unix')
receipts['pointsAwardedDate'] = pd.to_datetime(receipts['pointsAwardedDate'], errors='ignore', unit='ms', origin='unix')
receipts['purchaseDate'] = pd.to_datetime(receipts['purchaseDate'], errors='ignore', unit='ms', origin='unix')

# Explode the 'rewardsReceiptItemList' column from receipts dataset
receipts_xplo = receipts.explode('rewardsReceiptItemList')

# Insert empty dict for 'rewardsReceiptItemList' missing values (To use json_normalize function)
receipts_xplo['rewardsReceiptItemList'] = receipts_xplo['rewardsReceiptItemList'].apply(lambda x: {} if pd.isna(x) else x)

# Reset index & drop the old index column
receipts_xplo = receipts_xplo.reset_index()
receipts_xplo = receipts_xplo.drop(columns=['index'])

# Normalize JSON data into a flat table
itemlist = json_normalize(data=receipts_xplo['rewardsReceiptItemList'])

# Add reference column to itemlist dataset
itemlist = pd.concat([receipts_xplo['id'], itemlist], axis=1)
itemlist.rename(columns={'id': 'receipts_id'}, inplace=True)
```

```python
# Check for missing value
itemlist.isnull().sum()
```

```python
# Create csv files from the datasets (to import to MS SQL Server)
users.to_csv('D:/users/FR/users.csv', index=False)
brands.to_csv('D:/users/FR/brands.csv', index=False)
itemlist.to_csv('D:/users/FR/itemlist.csv', index=False)
receipts.to_csv('D:/users/FR/receipts.csv', index=False)
```
